```
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

def find_ball(img,color_range):
  img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
  r1_mask = cv2.inRange(img_hsv, color_range[0], color_range[1])
  r2_mask = cv2.inRange(img_hsv, color_range[2], color_range[3])
  mask = np.repeat((r1_mask | r2_mask)[:, :, np.newaxis], 3, axis=2)

  img_masked = img.copy()
  img_masked[mask == 0] = 0       
  _,contours, hierarchy	= cv2.findContours((r1_mask|r2_mask), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
  print('Have ' + str(len(contours)) + ' of red balls')    
  
red_ball_range = [(0,100,120), (10,255,255),(170,100,120), (180,255,255)] 
image = cv2.imread('lab2.jpg')
cv2_imshow(image)
find_ball(image,red_ball_range)
```
