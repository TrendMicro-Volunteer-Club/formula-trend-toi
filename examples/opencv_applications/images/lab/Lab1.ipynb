```
import cv2
from google.colab.patches import cv2_imshow

def find_egg(img,color_range):
  img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
  mask = cv2.inRange(img_hsv,color_range[0],color_range[1] )
  #cv2_imshow(mask)
  _, contours, hierarchy = cv2.findContours(mask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
  for cnt in contours:
    area = cv2.contourArea(cnt)     
    if area > 9000:               
      x, y, w, h = cv2.boundingRect(cnt) 
      cv2_imshow(img[y:y + h, x:x + w].copy())    
  
yellow_egg_range = [(20, 50, 50),(40, 255, 255)] 
image = cv2.imread('lab1.jpg')
cv2_imshow(image)
find_egg(image,yellow_egg_range)
```
